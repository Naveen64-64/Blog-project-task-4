# Task 3: Blog Application Notes

Enhance the existing Blog Application by adding **Search**, **Pagination**, and **UI Improvements**.

## 1. Search Functionality 🔍
- Add a search box.
- Search posts by **title** or **content**.
- Show all posts if search is empty.
- Display **'No posts found'** if there are no matches.
- Use SQL `LIKE` with prepared statements.

**Example SQL**
```sql
SELECT * FROM posts
WHERE title LIKE ? OR content LIKE ?;
```

## 2. Pagination 📄
- Show 5 or 10 posts per page.
- Use SQL `LIMIT` and `OFFSET`.
- Add Previous, Next, and page numbers.
- Highlight the current page.
- Keep pagination working with search.

**Example**
```sql
SELECT * FROM posts
LIMIT 5 OFFSET 10;
```

## 3. UI Improvements 🎨
- Use Bootstrap 5.
- Responsive layout.
- Better navbar, buttons, forms, spacing, and typography.
- Display posts as cards or a table.
- Add hover effects and improve user experience.
