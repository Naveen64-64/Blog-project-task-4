# Task 2: Basic CRUD Application (PHP + MySQL)

Simple explanation and roadmap.

## Objective
Create a blog website where users can Register, Login, Add Posts, View Posts, Edit Posts, Delete Posts, and Logout.

## Step 1: Create Database

````
CREATE DATABASE blog;

CREATE TABLE users(
    id INT AUTO_INCREMENT PRIMARY KEY,
    username VARCHAR(100),
    password VARCHAR(255)
);

CREATE TABLE posts(
    id INT AUTO_INCREMENT PRIMARY KEY,
    title VARCHAR(255),
    content TEXT,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
````

## Step 2: User Registration
Create `register.php` to store username and password.

## Step 3: User Login
Create `login.php` to verify users and start sessions.

## Step 4: Create Post
Create `add_post.php` to insert blog posts.

## Step 5: Read Posts
Create `index.php` to display all posts.

## Step 6: Update Post
Create `edit_post.php` to edit existing posts.

## Step 7: Delete Post
Create `delete_post.php` to remove posts.
